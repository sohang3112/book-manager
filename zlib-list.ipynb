```python
from typing import Dict
from pprint import pprint

import pandas as pd
import requests
from bs4 import BeautifulSoup

def view_html(html: str) -> bool:
    """
    View HTML in web browser
    """
    from tempfile import NamedTemporaryFile
    import webbrowser

    with NamedTemporaryFile(mode='w', suffix='.html', delete=False) as f:
        f.write(html)
        return webbrowser.open(f.name)

def text_and_url(link) -> Dict[str, str]:
    """
    @return: { 'text': str, 'url': str }
    """
    return { 'text': link.text, 'url': link.get('href') }
```

```python
query = 'seven'
res = requests.get(f'https://1lib.in/s/{query}')
assert res.status_code == 200
soup = BeautifulSoup(res.content, 'html5lib')
```

```python
books = soup.find_all('tr', class_='bookRow')
ans = [
    {
        'title': text_and_url(book.find('h3', itemprop='name').find('a')),
        'author': text_and_url(book.find('a', itemprop='author')),
    }
    for book in books
]
ans
```
